The chart is used to demonstrate how the number of visitors (in thousands) from three different web traffic sources — Searches, Direct, and Social Media — changed over five months (from 07/2019 to 11/2019).

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data
months = ['07/2019', '08/2019', '09/2019', '10/2019', '11/2019']
searches = [50, 53, 59, 56, 62]
direct = [39, 47, 42, 51, 51]
social_media = [70, 80, 90, 87, 92]
x = np.arange(len(months))  # label locations
width = 0.25  # width of the bars
```

```python
# Create plot
fig, ax = plt.subplots(figsize=(10,6))

# Bars
rects1 = ax.bar(x - width, searches, width, label='Searches', color='dodgerblue')
rects2 = ax.bar(x, direct, width, label='Direct', color='lightcoral')
rects3 = ax.bar(x + width, social_media, width, label='Social Media', color='orange')

# Labels & Title
ax.set_ylabel('visitors\nin thousands')
ax.set_xlabel('months')
ax.set_title('Visitors by web traffic sources')
ax.set_xticks(x)
ax.set_xticklabels(months)
ax.legend()

# Annotate bars
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height}',
                    xy=(rect.get_x() + rect.get_width()/2, height),
                    xytext=(0,3),
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

plt.tight_layout()
plt.show()
```
